```python
# Подключаем библиотеки и файлы для обработки, анализа и визуализации.

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import pandas as pd
import numpy as np
plt.style.use('default')
churn = pd.read_csv('/Users/dmitr/project_vs/data/data/churn.csv')
```

```python
# Задание 9.1.  Каково соотношение ушедших и лояльных клиентов? Покажите это на графике и дайте комментарий по соотношению.



customer_count = churn['Exited'].value_counts()

fig = plt.figure(figsize=(10,4))
main_axes = fig.add_axes([0,0,1,1])
main_axes.bar(x= customer_count.index, height = customer_count , color = ['blue' , 'red'])
main_axes.set_title('Соотношение ушедших и лояльных клиентов')
main_axes.set_xlabel('Статус клиента')
main_axes.set_ylabel('Количество клиентов')
main_axes.set_xticks([0, 1])
main_axes.set_xticklabels(['Лояльные', 'Ушедшие'])


loyal_count = customer_count[0]
churned_count = customer_count[1]
ratio = churned_count / loyal_count
print(f"Лояльные клиенты: {loyal_count}")
print(f"Ушедшие клиенты: {churned_count}")
print(f"Соотношение ушедших к лояльным: {ratio:.2f}")


'''
График показывает количество лояльных клиентов (слева) и количество ушедших клиентов (справа). В результате анализа было установлено:
Лояльные клиенты: 7963
Ушедшие клиенты: 2037
Соотношение ушедших к лояльным: 0.26
Таким образом, приблизительно 26% клиентов из общего числа ушли, что может указывать на необходимость анализа причин оттока и улучшения клиентского опыта для повышения лояльности.

'''
```

```python
#Задание 9.10 9.10. Переведите числовой признак CreditScore в категориальный. Для этого воспользуйтесь функцией get_credit_score_cat(), которая приведена ниже. Примените её к столбцу CreditScore и создайте новый признак CreditScoreCat — категории кредитного рейтинга.

#Постройте сводную таблицу, строками которой являются категории кредитного рейтинга (CreditScoreCat), а столбцами — количество лет, в течение которых клиент пользуется услугами банка (Tenure). В ячейках сводной таблицы должно находиться среднее по признаку оттока (Exited) — доля ушедших пользователей.

#На основе полученной сводной таблицы постройте тепловую карту с аннотацией. Найдите на тепловой карте категории клиентов, которые уходят чаще всего.

def get_credit_score_cat(credit_score):
    if credit_score >= 300 and credit_score < 500:
        return "Very_Poor"
    elif credit_score >= 500 and credit_score < 601:
        return "Poor"
    elif credit_score >= 601 and credit_score < 661:
        return "Fair"
    elif credit_score >= 661 and credit_score < 781:
        return "Good"
    elif credit_score >= 781 and credit_score < 851:
        return "Excellent"
    elif credit_score >= 851:
        return "Top"
    elif credit_score < 300:
        return "Deep"


churn['CreditScoreCat']  = churn['CreditScore'].apply(get_credit_score_cat)

pvt = churn.pivot_table(values= 'Exited' , index= 'CreditScoreCat' , columns='Tenure' , aggfunc= 'mean')

fig = plt.figure(figsize=(10,8))
sns.heatmap(pvt ,cmap='magma' , annot= True ,  fmt = '.2f' )
plt.title()

''' 
Самый высокий рейтинг оттока наблюдается у клиентов с очень плохим кредитным рейтингом.
Еще один из самых высоких показателей оттока наблюдается и у клиентов с длительностью пользования услугами 10 лет.
Соответсвенно можно сделать вывод что клиенты с низким кредитным рейтингом вероятнее покинут банк в первый год или в течении длительного пребывания. 
Клиенты с высоким кредитным рейтингом имеют меньшгший показатель оттока, а ввероятность ухода велика лишь в первый год.
Как основную стратегию по улучшению удержания клиентов, стоит более сосредоточится на поддержке и специальных программах для клиентов с низким кредитным рейтингом, особенно
в критические периоды их пребывания.
'''
```
